# Угадай число
Нужно написать программу, которая угадывает число за минимальное число попыток.
## Условия соревнования:
Компьютер загадывает целое число от 1 до 100, и нам его нужно угадать. Под «угадать», подразумевается «написать программу, которая угадывает число».
Алгоритм учитывает информацию о том, больше ли случайное число или меньше нужного нам.
Необходимо добиться того, чтобы программа угадывала число меньше, чем за 20 попыток.


```python
import numpy as np
import random
```

## Мое решение
Необходимо написать функцию, которая принимает на вход загаданное число и возвращает число попыток угадывания.
Цель - менее 20 попыток

```python
def game_core_v3(number: int = 1) -> int:
    """
    Args:
        number (int, optional): Загаданное число. Defaults to 1.

    Returns:
        int: Число попыток
    """
    # Ваш код начинается здесь

    count = 0
    bottom = 1  # присваиваем значение, соответствующее нижней границе диапазона поиска
    top = 101  # присваиваем значение, соответствующее верхней границе диапазона поиска +1

    while True:
        average_number = (bottom + top)//2
        count += 1
        if average_number == number:
            break  # выход из цикла если угадали
        elif number > average_number:
            bottom = average_number  # перезаписываем переменную bottom
        else:
            top = average_number # перезаписываем переменную top

    return count

    # Ваш код заканчивается здесь
```

Функция для оценки

```python
def score_game(random_predict) -> int:
    """За какое количество попыток в среднем за 10000 подходов угадывает наш алгоритм

    Args:
        random_predict ([type]): функция угадывания

    Returns:
        int: среднее количество попыток
    """
    count_ls = []
    #np.random.seed(1)  # фиксируем сид для воспроизводимости
    random_array = np.random.randint(1, 101, size=(10000))  # загадали список чисел

    for number in random_array:
        count_ls.append(random_predict(number))

    score = int(np.mean(count_ls))
    print(f"Ваш алгоритм угадывает число в среднем за: {score} попытки")
```

```python
print('Run benchmarking for game_core_v3: ', end='')
score_game(game_core_v3)
```
