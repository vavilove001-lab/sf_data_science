арифметические операции сохраняют NumPy-тип данных:

```python
a = np.int32(1000)
b = a + 25
print(b)
# 1025
print(type(b))
# <class 'numpy.int64'>
```

Если операция проводится с двумя NumPy-типами с фиксированным объёмом памяти, в результате сохраняется наиболее «старший» тип:

```python
a = np.int32(1000)
b = np.int8(25)
c = a + b
print(c)
# 1025
print(type(c))
```

если выделенной памяти для хранения переменной окажется недостаточно.

```python
a = np.int32(2147483610)
b = np.int32(2147483605)
print(a, b)

print(a + b)
```

```python
a = np.int32(2147483610)
b = np.int32(2147483605)
print(a, b)

print(np.int64(a) + np.int64(b))
```

```python
a = True
print(type(a))
# <class 'bool'>
a = np.bool(a)
print(type(a))
# <class 'bool'>
a = np.bool_(a)
print(type(a))
# <class 'numpy.bool_'>
 
# Значения равны
print(np.bool(True) == np.bool_(True))
# True
# А типы — нет:
print(type(np.bool(True)) == type(np.bool_(True)))
# False
```
